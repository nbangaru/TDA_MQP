```python
# import packages

from itertools import product

import time
import pandas as pd
import numpy as np
from sklearn import datasets
from scipy.stats import multivariate_normal as mvn
import matplotlib.pyplot as plt

import teaspoon.MakeData.PointCloud as makePtCloud

from ripser import Rips
from persim import PersistenceImager
from sklearn_extra.cluster import KMedoids
from persim.images_weights import linear_ramp
```

```python
# define rips object
rips = Rips()

#list will store all H1 diagrams
alldgms = []

#set parameters for persistence images
pimgr.pixel_size = 0.1
pimgr.weight = linear_ramp
pimgr.weight_params = {'low':0.0, 'high':1.0, 'start':0.0, 'end':10.0}
```

```python
#2024-2025 

#import and convert data to matrix
sheet_id = '1Mkqw6vHcfSs18qN6kd7X7TDp69KhR37Dqgmi9Fcrmjk'
df = pd.read_csv(f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv")
dataMatrix = df.to_numpy()

#get H0 and H1 Diagrams
dgms = rips.fit_transform(dataMatrix)
H0_dgm = dgms[0]
H1_dgm = dgms[1]

#plot H0 and H1 Diagrams
plt.subplot(122)
rips.plot(dgms, legend=False, show=False)
plt.title("2024-2025 Persistence Diagram of $H_0$ and $H_1$")
plt.show()

#add H1 diagram to dgms list
alldgms.append(H1_dgm)

#plot persistence image
pimgr.fit(H1_dgm)
fig, axs = plt.subplots(1, 2, figsize=(20,5))
pimgr.plot_diagram(H1_dgm, skew=True, ax=axs[0])
axs[0].set_title('Diagram', fontsize=16)
pimgr.plot_image(pimgr.transform(H1_dgm), ax=axs[1])
axs[1].set_title('Pixel Size: 0.1', fontsize=16)
plt.tight_layout()

#check length of dgms
print("diagrams length: ", len(alldgms))
```

```python
vecs = [pimgr.transform(diagram).flatten() for diagram in alldgms]
X = np.vstack(vecs)  

kmedoids = KMedoids(n_clusters=3, random_state=42, metric='euclidean')
labels = kmedoids.fit_predict(X)

numberOfSeasons = len(labels)
numberOfClusters = len(kmedoids.medoid_indices_)


#printing data neatly
for i in range(numberOfClusters): 
    cluster = []
    for j in range(numberOfSeasons):
        if(labels[j] == i):
            cluster.append(2024-j)
    clusterSize = len(cluster)
    print()
    print("Cluster ", i, " Mediod: ", 2024-kmedoids.medoid_indices_[i], "-", 2024-kmedoids.medoid_indices_[i]+1)
    print("Cluster ", i, ":")
    for k in range(clusterSize):
        print(cluster[k],"-",cluster[k]+1)


```
